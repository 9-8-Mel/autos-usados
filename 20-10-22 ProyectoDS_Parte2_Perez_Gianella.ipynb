## 6.14. Precios de ventas según tipo de combustibles:
---


```python
groups = df[['fuel','selling_price']].groupby('fuel')
q1 = groups.quantile(q=0.25)
q2 = groups.quantile(q=0.5)
q3 = groups.quantile(q=0.75)

iqr = q3 - q1

upper = q3 + 1.5*iqr
lower = q1 - 1.5*iqr

def outliers(group):
    fuel = group.name 
    return group[(group.selling_price > upper.loc[fuel]['selling_price']) | (group.selling_price < lower.loc[fuel]['selling_price'])]['selling_price']

out = groups.apply(outliers).dropna()  

if not out.empty: 
    outx = list(out.index.get_level_values(0)) 
    outy = list(out.values)

cats= ['Petrol','Diesel','CNG']

p = figure(background_fill_color="#efefef", x_range=cats, title='Comparacion precio con tipo de combustible')

qmin = groups.quantile(q=0.00)
qmax = groups.quantile(q=1.00)

upper['score'] = [min([x,y]) for (x,y) in zip(list(qmax.loc[:,'selling_price']),upper.selling_price)]
lower['score'] = [max([x,y]) for (x,y) in zip(list(qmin.loc[:,'selling_price']),lower.selling_price)]

p.segment(cats, upper.selling_price, cats, q3.selling_price, line_color="black")
p.segment(cats, lower.selling_price, cats, q1.selling_price, line_color="black")

p.vbar(cats, 0.7, q2.selling_price, q3.selling_price, fill_color="#E08E79", line_color="black") 
p.vbar(cats, 0.7, q1.selling_price, q2.selling_price, fill_color="#3B8686", line_color="black")

p.rect(cats, lower.selling_price, 0.2, 0.01, line_color="black")
p.rect(cats, upper.selling_price, 0.2, 0.01, line_color="black")

if not out.empty:
    p.circle(outx, outy, size=6, color="#F38630", fill_alpha=0.6)

p.ygrid.grid_line_color = "white"
p.grid.grid_line_width = 2
p.xaxis.major_label_text_font_size="16px"
p.xaxis.axis_label ='Combustible'
p.yaxis.axis_label='Precio (u$s)'
p.add_tools(HoverTool())

show(p)
```
